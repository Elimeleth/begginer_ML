```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch.nn.functional as F
import torch  
```

```python
model_name = 'distilbert-base-uncased-finetuned-sst-2-english'
```

```python
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
X_TRAIN = ["we are happy and we have good news", "you are fat person", "you are sick, stupid"]
```

```python
tokens = tokenizer.tokenize("we are happy but we dont have good news")
tokens_ids = tokenizer.convert_tokens_to_ids(tokens)
inputs_ids = tokenizer("we are happy but we dont have good news")
```

```python
batch = tokenizer(X_TRAIN, padding=True, max_length=512, truncation=True, return_tensors='pt')
```

```python
with torch.no_grad():
  out = model(**batch, labels=torch.tensor([1,0, 0]))
  print(out)
  predictions = F.softmax(out.logits, dim=1)
  print(predictions)
  labels = torch.argmax(predictions, dim=1)
  print(labels)
  labels = [model.config.id2label[label_id] for label_id in labels.tolist()]
  print(labels)
```
